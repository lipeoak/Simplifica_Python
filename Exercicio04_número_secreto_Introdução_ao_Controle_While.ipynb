Vamos criar um programa que pede ao usu√°rio para adivinhar um n√∫mero aleat√≥rio e fornece dicas sobre se o palpite √© muito alto ou muito baixo. O loop while continuar√° at√© que o usu√°rio adivinhe corretamente o n√∫mero.

```python
import random

numero_secreto = random.randint(1,100)
palpite = 0

print("Descubra o n√∫mero secreto!!!")

while palpite != numero_secreto:
  palpite = int(input("Informe um n√∫mero entre 1 e 100 para tentar acertar o n√∫mero secreto: "))
  if palpite > numero_secreto:
    print(f"{palpite} √© maior que o n√∫mero secreto")
  elif palpite < numero_secreto:
    print(f'{palpite} √© menor que o n√∫mero secreto')
print(f'Voc√™ acertou o n√∫mero secreto! O n√∫mero secreto √©: {palpite}')
```

ü§ñ **Estruturas de Controle (while)**

A estrutura de controle **while** √© utilizada para criar loops, ou seja, repetir um bloco de c√≥digo enquanto uma condi√ß√£o espec√≠fica for verdadeira.

while.jpg


A sintaxe b√°sica do loop while em Python √© a seguinte:

```python
while condi√ß√£o:
    # C√≥digo a ser executado enquanto a condi√ß√£o for verdadeira
    # ...
```

A condi√ß√£o √© uma express√£o booleana que √© avaliada antes de cada itera√ß√£o do loop. Enquanto a condi√ß√£o for verdadeira, o bloco de c√≥digo dentro do loop √© executado. Assim que a condi√ß√£o se torna falsa, a execu√ß√£o do loop √© interrompida, e o programa continua com a pr√≥xima instru√ß√£o ap√≥s o loop.

Aqui est√° um exemplo simples de um loop while que imprime os n√∫meros de 1 a 5:

```python
numero = 4

while numero <=5:
  print(numero)
  numero = numero +1

print("Terminou o programa")

```

Neste exemplo, o loop continuar√° a ser executado enquanto a vari√°vel contador for menor ou igual a 5. A cada itera√ß√£o, o valor de contador √© impresso e incrementado em 1. O loop ser√° interrompido quando contador se tornar 6, pois a condi√ß√£o contador <= 5 se tornar√° falsa.

√â importante ter cuidado ao usar loops while para evitar loops infinitos. Certifique-se de que a condi√ß√£o do loop eventualmente se torne falsa para que o loop possa ser encerrado. Caso contr√°rio, o programa ficar√° preso em um loop infinito, o que pode levar a problemas de desempenho e at√© mesmo √† parada for√ßada do programa.

```python
import random

numero_secreto = random.randint(1, 100)

print ("Bem-vindo ao programa ADIVINHA")
print ("Tente adivinhar o numero secreto")

palpite = 0

while palpite != numero_secreto:
  palpite = int(input("Digite o seu palpite:"))

  if palpite  > numero_secreto:
    print("Seu palpite est√° acima do numero secreto")
  elif palpite  < numero_secreto:
    print("Seu palpite est√° abaixo do numero secreto")
  else:
    print(f"Voc√™ acertou o numero secreto: {numero_secreto}")

```

Neste exemplo, o programa gera um n√∫mero secreto aleat√≥rio entre 1 e 100. O usu√°rio √© solicitado a fornecer palpites, e o programa fornece dicas sobre se o palpite √© muito alto, muito baixo ou correto. O loop while continua at√© que o usu√°rio adivinhe corretamente o n√∫mero secreto.
